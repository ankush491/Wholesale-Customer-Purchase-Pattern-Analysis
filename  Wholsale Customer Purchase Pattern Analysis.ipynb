# From above boxplot we get to know that , there are many outliers, So we should remove it.

```python
for feature in var_columns:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1-1.5*IQR
    upper_fence = Q3+1.5*IQR
    for i in range(len(df)):
        outlier = df[feature][i]
        if outlier > upper_fence:
            df[feature][i] = upper_fence
        elif outlier < lower_fence:
            df[feature][i] = lower_fence
```

```python
plot_number = 1
plt.figure(figsize = (15,10))
for i in var_columns:
    plt.subplot(3,2,plot_number)
    sns.boxplot(y=df[i]);
    plot_number+=1
```

# XGBoost Algorithm

```python
from xgboost import XGBClassifier
XG=XGBClassifier()
```

```python
# Map the 'Channel' column to 0 and 1
df['Channel'] = df['Channel'].map({1: 0, 2: 1})
```

```python
# Separate features (x) and target (y)
x = df.drop("Channel", axis=1)
y = df.Channel
```

```python
# Scale the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=20)
```

```python
# Import and train the XGBoost classifier
from xgboost import XGBClassifier
XG = XGBClassifier()
XG.fit(x_train, y_train)
```

```python
y_predict_XG=XG.predict(x_test)
y_predict_XG

```

```python
accuracy_score(y_test,y_predict_XG)
```

```python
r2_score(y_test,y_predict_XG)
```

# Performing Cross Validation and Resulting the Scores

```python
from sklearn.model_selection import cross_val_score
clf = [knn,SV,dst,RC,XG]

for algo in clf:
    score = cross_val_score( algo,x,y,cv = 10,scoring = 'accuracy')
    print("The accuracy score of {} is:".format(algo),score.mean())
```

# Using Stack Model

```python
clf = [('knn',knn),('SV',SV),('DST',dst),('xgb',XG),('RC',RC)]
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
stack_model = StackingClassifier( estimators = clf,final_estimator = lr)
```

```python
score = cross_val_score(stack_model,x,y,cv = 5,scoring = 'accuracy')
```

```python
print("The accuracy score of is:",score.mean())
```
